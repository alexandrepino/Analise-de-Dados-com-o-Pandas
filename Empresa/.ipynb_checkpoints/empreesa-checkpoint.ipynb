```python
import pandas as pd
import matplotlib.pyplot as plt
```

#  Lendo os arquivos CSV com o pandas

```python
funcionario_df = pd.read_csv('CadastroFuncionarios.csv', sep =";")
cliente_df = pd.read_csv('CadastroClientes.csv', sep = ";")
servico_df = pd.read_excel('BaseServiçosPrestados.xlsx')




```

# Formatando e Convertendo a tabela Funcionarios

```python
funcionario_df['VR'] = funcionario_df['VR'].apply(lambda x: str(x).replace(',','.'))
funcionario_df['Beneficios'] = funcionario_df['Beneficios'].apply(lambda x: str(x).replace(',','.'))
funcionario_df['Impostos'] = funcionario_df['Impostos'].apply(lambda x: str(x).replace(',','.'))
funcionario_df[['Salario Base','VT', 'Beneficios', 'VR', 'Impostos']] = funcionario_df[['Salario Base','VT', 'Beneficios', 'VR', 'Impostos']].astype(float)
pd.set_option('float_format','{:.2f}'.format)


funcionario_df



```

# Formatando e Convertendo a tabela de Clientes

```python
cliente_df['Valor Contrato Mensal'] = cliente_df['Valor Contrato Mensal'].astype(float)
pd.set_option('float_format','{:.2f}'.format)

cliente_df.head(10)
```

# Formatando e Convertendo a tabela Funcionarios

```python
servico_df.head(10)
```

# Calculando o Salário Bruto de cada Funcionário

```python
funcionario_df['Salario Bruto'] = funcionario_df['VR'] + funcionario_df['Impostos'] + funcionario_df['VT'] + funcionario_df['Beneficios'] + funcionario_df['Salario Base']
funcionario_df
```

# Calculando o Total gasto com Salário

```python
despesa_salarios = funcionario_df['Salario Bruto'].sum()
print('O total gasto com salario de funcionários é R${:,}'.format(despesa_salarios))
```

# Calculando o faturamento total da Empresa

```python
faturamento_df = servico_df[['ID Cliente', 'Tempo Total de Contrato (Meses)']].merge(cliente_df[['ID Cliente','Valor Contrato Mensal']], on='ID Cliente')
faturamento_df['Faturamento Total'] = faturamento_df['Tempo Total de Contrato (Meses)'] * faturamento_df['Valor Contrato Mensal']
faturamento_df
```

```python
faturamento_total = faturamento_df['Faturamento Total'].sum()
print('O faturamento total da empresa foi de R${:,}'.format(faturamento_total))
```

# Porcentagem de funcionarios que assinaram pelo menos um contrato

```python
total_funcionarios = len(funcionario_df['ID Funcionário'])
total_funcionarios_que_venderam = len(servico_df['ID Funcionário'].drop_duplicates())
porcentagem = total_funcionarios_que_venderam / total_funcionarios
print('A porcentagem de funcionário que venderam foi de {:.2%}'.format(porcentagem))
```

# Contratos fechados por Area

```python
contratos_por_area = servico_df[['ID Funcionário']].merge(funcionario_df[['ID Funcionário', 'Area']], on = 'ID Funcionário')
contratos = contratos_por_area["Area"].value_counts()
print(contratos)
contratos.plot(kind = 'bar')
```

# Numero de Funcionarios por Area

```python
funcionario_por_area = funcionario_df['Area'].value_counts()
print(funcionario_por_area)
funcionario_por_area.plot(kind = 'bar')
```

# Calculando a média do ticket dos Clientes

```python
media_ticket = cliente_df['Valor Contrato Mensal'].mean().round(2)
print("O Ticket médio é de R${:,}".format(media_ticket))


```
