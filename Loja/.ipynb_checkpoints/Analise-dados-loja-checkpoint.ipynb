# Importando o pacote pandas


```python
import pandas as pd
```

#  Lendo os arquivos CSV com o pandas

```python
vendas_df = pd.read_csv('Contoso-Vendas-2017.csv', sep = ';')
produtos_df = pd.read_csv('Contoso - Cadastro Produtos.csv', sep = ';')
lojas_df = pd.read_csv('Contoso - Lojas.csv', sep =';')
clientes_df = pd.read_csv('Contoso - Clientes.csv', sep = ';')
produtos_df
```

# Convertendo uma coluna de object para datetime

```python
vendas_df['Data da Venda'] = pd.to_datetime(vendas_df['Data da Venda'], format='%d/%m/%Y')
vendas_df['Data do Envio'] = pd.to_datetime(vendas_df['Data do Envio'], format='%d/%m/%Y')

vendas_df.info()
```

#  Alterando ó preço do produto com ID 873 para 23 reais


```python
produtos_df.loc[produtos_df["ID Produto"] == 873, "Preco Unitario"] = 23.00
produtos_df
```

#  Atualizando e exportando o arquivo csv

```python
produtos_df.to_csv('Novo Produtosss.csv', sep = ';')
```
